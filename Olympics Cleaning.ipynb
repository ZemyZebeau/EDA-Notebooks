I noticed a number of cases within athletes' "Event" column had an inconsistent case format. I Googled how to address this and these functions were returned.

```python
def apostrophe(text):
    words = text.split()
    new_words = []
    for word in words:
        if "-" in word:
            parts = word.split("-")
            new_word = parts[0].capitalize() + "-" + parts[1]
        else:
            new_word = word.capitalize()
        new_words.append(new_word)
    return " ".join(new_words)


def hyphens(text):
    words = text.split()
    capitalized_words = []
    for word in words:
        if '-' in word:
            hyphenated_words = word.split('-')
            capitalized_hyphenated_words = [w.capitalize() for w in hyphenated_words]
            capitalized_words.append('-'.join(capitalized_hyphenated_words))
        else:
            capitalized_words.append(word.capitalize())
    return ' '.join(capitalized_words)


import re

def capitalize_after_paren(text):
    return re.sub(r'\(([\w\s]+)', lambda match: '(' + ' '.join(word.capitalize() for word in match.group(1).split()), text)
```

```python
cleaned_athletes.Event = cleaned_athletes.Event.apply(apostrophe)
cleaned_athletes.Event = cleaned_athletes.Event.apply(hyphens)
cleaned_athletes.Event = cleaned_athletes.Event.apply(capitalize_after_paren)
```

I discovered the functions missed a few cases with double hyphens and and apostrophes. I used .replace() to address these cases.

```python
cleaned_athletes = cleaned_athletes.replace("Luge Mixed (Men)'s Doubles", "Luge Mixed Men's Doubles")
cleaned_athletes = cleaned_athletes.replace("Tug-Of Men's Tug-Of", "Tug-Of-War Men's Tug-Of-War")
```
