## File Merge

```python
df_noc = pd.read_csv(r'/content/drive/MyDrive/Newman/Descriptive Analytics/Olympics EDA /data/noc_regions.csv', sep = ',', header = 'infer')
df_athletes = pd.read_csv(r'/content/drive/MyDrive/Newman/Descriptive Analytics/Olympics EDA /data/athlete_events.csv', sep =',', header = 'infer')
```

```python
# Creates nonull dataframe
nonull = df_noc[pd.notnull(df_noc['notes'])]
```

```python
# Creates df_joinedtable merging on nonull and df_athletes on NOC
df_joinedtable = nonull.merge(df_athletes, on='NOC', how='left')
```

```python
# Renames 'notes' and 'region' columns.
df_reorderedtable = df_joinedtable.loc[:, ['ID', 'Year', 'Name', 'Sex', 'Age', 'Height',	'Weight',	'NOC', 'Games', 'Season',
                                           'City', 'region',	'notes',	'Sport', 'Event', 'Medal']].rename(columns = {'notes':'Notes',
                                            'region':'Region'})
```

```python
# Sorts dataframe by index.
df_reorderedtable = df_reorderedtable.sort_index()
```

```python
male = df_reorderedtable[df_reorderedtable['Sex'] == 'M']
female = df_reorderedtable[df_reorderedtable['Sex'] == 'F']
```

```python
male = male['ID'].count()
female = female['ID'].count()
gentotal = male + female
male_percent = (male / gentotal)
fem_percent = (female / gentotal)
```

```python
data = pd.DataFrame([["Male", male, round(male_percent * 100)],["Female", female, round(fem_percent * 100)],["Total", gentotal, 100.0]])
data.columns = ['Sex', 'Count', 'Percent']
```
