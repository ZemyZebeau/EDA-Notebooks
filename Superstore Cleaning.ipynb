## Merge


```python
df_orders = pd.read_excel('/content/drive/MyDrive/Newman/Descriptive Analytics/Superstore EDA/data/Global_Superstore_Orders_2012_2015.xlsx', sheet_name = 0)
df_people = pd.read_excel('/content/drive/MyDrive/Newman/Descriptive Analytics/Superstore EDA/data/Global_Superstore_Orders_2012_2015.xlsx', sheet_name = 1)
```

```python
# print to grid function
def print_grid(data, width):
    """Prints a list in a grid format with a specified width."""
    for i in range(0, len(data), width):
        print(data[i:i + width])
    print()
```

```python
# places table headers in lists and runs them through the print to grid function
order_list = df_orders.columns.tolist()
people_list = df_people.columns.tolist()
print_grid(order_list, 5)
print_grid(people_list, 5)
```

```python
# merges (left) df_orders and df_people based on 'Region' columns.
df_mergetable = pd.merge(df_orders, df_people, left_on='Region', right_on = 'Region', how='left')
```

```python
# returns mergetable's headers
merge_list = df_mergetable.columns.tolist()
print_grid(merge_list, 5)
```

## Missing Values

I turned my attention to null values. Only the Postal Code column had any. I used isnull() to confirm.

```python
df_orders[df_orders.isnull().any(axis=1)]
```

*he Superstore operates internationally but only logs postal codes for customers in the US hence the large number of null fields. Furthermore, the column's data type is float resulting in postal codes with decimal points.

I decided to fill the null values of Postal code with zeros and change the data type from float to int to remove the decmial points. First I assigned the dataframe to a new variable "cleaned_orders". Then I filled the null fields with zeros to avoid errors. Finally I changed the datatype to int using ".astype()".

```python
cleaned_orders = df_orders
cleaned_orders['Postal Code'] = cleaned_orders['Postal Code'].fillna(0)
cleaned_orders['Postal Code'] = cleaned_orders['Postal Code'].astype(int)
```
